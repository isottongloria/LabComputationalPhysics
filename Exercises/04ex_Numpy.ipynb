7\. Prime numbers sieve: compute the prime numbers in the 0-N (N=99 to start with) range with a sieve (mask).
  * Constract a shape (100,) boolean array, the mask
  * Identify the multiples of each number starting from 2 and set accordingly the corresponding mask element
  * Apply the mask to obtain an array of ordered prime numbers
  * Check the performances (timeit); how does it scale with N?
  * Implement the optimization suggested in the [sieve of Eratosthenes](https://en.wikipedia.org/wiki/Sieve_of_Eratosthenes)

````python
#a
mask = np.array([True for i in range(101)])
print(mask)

for num in range(2,100,1):
    for j in range(2, num):
        if (num % j) == 0:
            mask[num]=bool(False)
            break
                
#set value for num=1,2, 3
mask[0]=False  # 0 is not a prime
mask[1]=False #1 is not prime
mask[2]=True # 2,3 areprime
mask[3]=True 

print(mask) 

#c
prime = np.array([i for i in range(101) if mask[i]])
print(prime)

#d sieve of Eratosthenes
n=99
prime = [True for i in range(n+1)]
p = 2
while (p * p <= n):
 
    # If prime[p] is not
    # changed, then it is a prime
    if (prime[p] == True):
 
        # Update all multiples of p
        for i in range(p * p, n+1, p):
            prime[i] = False
    p += 1
 
    # Print all prime numbers
for p in range(2, n+1):
    if prime[p]:
        print(p)
````

8\. Diffusion using random walk

Consider a simple random walk process: at each step in time, a walker jumps right or left (+1 or -1) with equal probability. The goal is to find the typical distance from the origin of a random walker after a given amount of time. 
To do that, let's simulate many walkers and create a 2D array with each walker as a raw and the actual time evolution as columns

  * Take 1000 walkers and let them walk for 200 steps
  * Use randint to create a 2D array of size walkers x steps with values -1 or 1
  * Build the actual walking distances for each walker (i.e. another 2D array "summing on each raw")
  * Take the square of that 2D array (elementwise)
  * Compute the mean of the squared distances at each step (i.e. the mean along the columns)
  * Plot the average distances (sqrt(distance\*\*2)) as a function of time (step)
  
Did you get what you expected?

````python
import random

def rnd():
    x=random.randint(0,1)
    if x==0:
        x=-1
    return x

walkers=np.zeros((1000,200))
for i in range(1000):
    for j in range(200):
        walkers[i,j]=rnd()
print(walkers)

summing_each_row=np.zeros((1000,200)) #built the actual (=cumulative) walking distance
for i in range(1000):  # iterate through rows of walkers
    for j in range(200): # iterate through columns of walkers
        for k in range(j):
            summing_each_row[i][j]+=walkers[i][k]
print(summing_each_row)


walkers_sum_sqr = summing_each_row**2
print(sum(walkers_sum_sqr))
mean = np.sqrt(sum(walkers_sum_sqr)/1000)

time = [i for i in range(200)]

import matplotlib.pyplot as plt
plt.plot(time,mean)
plt.xlabel('Time')
plt.ylabel('Mean of the squared distances at each step')
plt.show() 
````

````python
#Load the data into a 2D array
file=np.loadtxt("populations.txt")
print(file)

#Create arrays out of the columns, the arrays being (in order): *year*, *hares*, *lynxes*, *carrots* 
year=[file[x,0] for x in range(file.shape[0])]
hares=[file[x,1] for x in range(file.shape[0])]
lynxes=[file[x,2] for x in range(file.shape[0])]
carrots=[file[x,3] for x in range(file.shape[0])]
#carrots=file[:,3] <----- metodo alternativo per selezionare le colonne

#Plot the 3 populations over the years
import matplotlib.pyplot as plt
plt.plot(year, hares,color='r', label='hares')
plt.plot(year, lynxes, color='b', label='lynxes')
plt.plot(year, carrots, color='g', label='carrots')
plt.xlabel('year')
plt.ylabel('populations')
plt.legend()
plt.show()

#Compute the main statistical properties of the dataset (mean, std, correlations, etc.)
print("media hares: " , np.mean(hares), "dev std hares:", np.std(hares), "corr hares-lynxes", np.corrcoef(hares,lynxes))
print("media lynxes: " , np.mean(lynxes), "dev std lynxes:", np.std(lynxes), "corr carrots-lynxes", np.corrcoef(carrots,lynxes))
print("media carrots: " , np.mean(carrots), "dev std carrots:", np.std(carrots), "corr hares-carrots", np.corrcoef(hares,carrots))
print('correlation: ',np.corrcoef([hares, lynxes,carrots]))

#Which species has the highest population each year? SEE ALSO NEXT INPUT BOX
highest_population=np.amax(file, axis=1)  #select the max number of individuals for each row

index=file.argmax(axis=1) #select the index(coloumn) of the max population each year
lista=[]                                    # and then  translate it in a list 1=hares, 2=lynxes, 3=carrots
for x in range(0,21,1):
    if index[x]==1:
        lista.append('hares')
    elif index[x]==2:
        lista.append('lynxes')
    elif index[x]==3:
        lista.append('carrots')
    
print(index)
for x in range(0,21,1):
    print(lista[x], highest_population[x])
````
